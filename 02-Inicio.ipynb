## Obtener los datos para los labs
Los archivos se encuentran en google drive

```python
import requests
from tqdm.notebook import tqdm

drive_files = [
    {
        'id':'1X0Cy_v_ayQZjtz-JaWx4eYyb30DOSvH-',
        'name': 'persona.data'
    },
    {
        'id':'1Jp-AgClWeD2ypLueNfHMuLQIUIE3MVGi',
        'name': 'transacciones.json'
    },
    {
        'id':'1f-At0wsEkn3qHGRyoXLYFb1CVyHAIi_b',
        'name': '2015-summary.csv'
    },
    {
        'id':'10atuc7Ke_to5NMfcudT2LCjhbk1Tgf3E',
        'name': 'transacciones.xml'
    }
]

for file in tqdm(drive_files, desc='Downloading files'):
    url = f'https://drive.google.com/uc?export=download&id={file['id']}'
    volume_path = f'/Volumes/platzi/customers/example_data/{file['name']}'

    response = requests.get(url)

    with open(volume_path, "wb") as f:
        f.write(response.content)
```

```python
dataframe = spark.read.csv('/Volumes/platzi/customers/example_data/2015-summary.csv',
                      header=True,
                      inferSchema=True)
dataframe.createOrReplaceTempView('2025summary')
```

```python
query = '''
    SELECT * 
    FROM 2025summary
    LIMIT 10
'''
spark.sql(query).show(truncate=False)
```

```python
dataframe.write.saveAsTable('platzi.customers.summary',
                            mode='overwrite')
```
