```python
#Lectura desde archivo texto plano: csv
summary_df = spark.read.csv('/Volumes/platzi/customers/example_data/2015-summary.csv', header="true", inferSchema="true")
```

```python
#Lectura desde archivo de texto plano: data
persona_df = spark.read.format("csv").option("header", "true").option("delimiter", "|").option("encoding", "ISO-8859-1").load("/Volumes/platzi/customers/example_data/persona.data")
```

```python
#Esquema de metadatos
#Importante todos los campos fueron reconocidos con el tipo de dato "string"
persona_df.printSchema()
```
